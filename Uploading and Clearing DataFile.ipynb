# Czyścimy oraz łączymy wgrane pliki csv!

## Zrobienie indeksu z kolumny "Zawodnik"

### Tworzymy dataFrame, ktory zawiera wszystkie kolumny, które posiadają wszystkie inne kolumny, czyli "Zawodnik","Drużyna" oraz "Mecze"

```python
mecze = punkty[["Zawodnik","Drużyna","Mecze"]]
mecze.set_index(mecze["Zawodnik"], inplace=True)
mecze = mecze.drop("Zawodnik", axis=1, inplace=False)
mecze.head()
```

```python
zbior_plikow = [asysty,bloki,czas_gry,eval,faule,przechwyty,punkty,rzuty_wolne,rzuty_z_gry,rzuty_za_2,rzuty_za_3,straty,zbiorki]

i = 0
while i<len(zbior_plikow):
    zbior_plikow[i].set_index(zbior_plikow[i]["Zawodnik"], inplace=True)
    zbior_plikow[i] = zbior_plikow[i].drop(["Lp.","Zawodnik","Drużyna","Mecze"], axis=1, inplace=True)
    i+=1
bloki
```

```python
# Łączenie dataframes
```

```python
df = pd.concat([mecze,czas_gry,punkty,rzuty_z_gry,rzuty_za_2,rzuty_za_3,rzuty_wolne,zbiorki,asysty,przechwyty,bloki,faule], axis=1, ignore_index=False)
df
```
